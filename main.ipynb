```python
import pygame
import time
from time import strftime
from time import gmtime

# window parameters
WIN_WIDTH = 1024
WIN_HEIGHT = 600

# button parameters
BTN_WIDTH = 80
BTN_HEIGHT = 80
BTN_GAP = 70

# HP parameters
HP_WIDTH = 40
HP_HEIGHT = 40
HP_GAP = 40
HP_yLOCATION = 0
HP_xLOCATION = 400
FPS = 30

# enemy parameters
ENEMY_xLOCATION = 40
ENEMY_yLOCATION = 265
ENEMY_xHP = 50
ENEMY_yHP = 4
ENEMY_WIDTH = 50
ENEMY_HEIGHT = 50

# menu parameters
MENU_HEIGHT = 80

# color (RGB)
BLACK = (0, 0, 0)
WHITE = (255, 255, 255)
RED = (255, 0, 0)

# timer parameters
TIMER_X = 60
TIMER_Y = 30

# initialization
pygame.init()

# load image (background, enemy, buttons)
background_image = pygame.transform.scale(pygame.image.load("images/Map.png"), (WIN_WIDTH, WIN_HEIGHT))
enemy_image = pygame.transform.scale(pygame.image.load("images/enemy.png"), (ENEMY_WIDTH, ENEMY_HEIGHT)) 
muse_image = pygame.transform.scale(pygame.image.load("images/muse.png"), (BTN_WIDTH, BTN_HEIGHT)) 
pause_image = pygame.transform.scale(pygame.image.load("images/pause.png"), (BTN_WIDTH, BTN_HEIGHT)) 
sound_image = pygame.transform.scale(pygame.image.load("images/sound.png"), (BTN_WIDTH, BTN_HEIGHT)) 
continue_image = pygame.transform.scale(pygame.image.load("images/continue.png"), (BTN_WIDTH, BTN_HEIGHT)) 
hp_image = pygame.transform.scale(pygame.image.load("images/hp.png"), (HP_WIDTH, HP_HEIGHT)) 
hp_gray_image = pygame.transform.scale(pygame.image.load("images/hp_gray.png"), (HP_WIDTH, HP_HEIGHT)) 

# create window surface
window = pygame.display.set_mode((WIN_WIDTH, WIN_HEIGHT))
# set window caption
pygame.display.set_caption("My first game")
# set font
FONT = pygame.font.SysFont("Arial", 30)
# clock
clock = pygame.time.Clock()



class Game:
    def __init__(self):
        # window
        self.window = pygame.display.set_mode((WIN_WIDTH, WIN_HEIGHT))
        # hp
        self.hp = 7
        self.max_hp = 10
        pass

        # game loop
    def game_run(self):
        # get starting time
        time_start = time.time() 
        run = True
        while run:
            clock.tick(FPS)
            # event loop (user action)
            for event in pygame.event.get():
                if event.type == pygame.QUIT:
                    run = False
            
            # how long did time go through
            time_count = time.time() - time_start

            # draw background
            self.window.blit(background_image,(0, 0))

            # draw enemy and health bar
            self.window.blit(enemy_image, (ENEMY_xLOCATION, ENEMY_yLOCATION))
            pygame.draw.rect(self.window, RED, [ENEMY_xLOCATION, ENEMY_yLOCATION, ENEMY_xHP, ENEMY_yHP])

            # draw menu (and buttons)
            pygame.draw.rect(self.window, BLACK, [0, 0, WIN_WIDTH, MENU_HEIGHT])
            self.window.blit(pause_image,(WIN_WIDTH-BTN_GAP, 0))
            self.window.blit(continue_image,(WIN_WIDTH-BTN_GAP*2, 0))
            self.window.blit(sound_image,(WIN_WIDTH-BTN_GAP*3, 0))
            self.window.blit(muse_image,(WIN_WIDTH-BTN_GAP*4, 0))
            
            # draw HP
            for i in range (0, self.max_hp):
                # 5 HP per row
                if i < 5:
                    j = 0
                    second_line = 0    
                else:
                    j = 5
                    second_line = 40
                # display 10 gray HP, and then use 7 green HP overwrite on it
                self.window.blit(hp_gray_image, (HP_xLOCATION + HP_WIDTH * (i - j), HP_yLOCATION + second_line))
                if i < self.hp :
                    self.window.blit(hp_image, (HP_xLOCATION + HP_WIDTH * (i - j), HP_yLOCATION + second_line))
            
            """time display"""
            # draw a black rect timer frame
            pygame.draw.rect(self.window, BLACK, [0, WIN_HEIGHT - 45, 100, 45])
            # time = current time - start time
            time_last = int(time.time() - time_start)
            # get time and turn into surface with white text color and black background
            time_surface = FONT.render( strftime("%M:%S", gmtime(time_last) ), True, WHITE, BLACK )
            self.window.blit(time_surface, (20, WIN_HEIGHT - 40))
            
            pygame.display.update()
        pygame.quit()

if __name__ == "__main__":
    covid_game = Game()
    covid_game.game_run()


```
